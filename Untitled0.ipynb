```python
import sqlite3
database='database.sqlite'
conn=sqlite3.connect(database)
print("opened data sucessfully")
```

```python
import pandas as pd
query="SELECT * FROM sqlite_master WHERE type='table'"
table=pd.read_sql_query(query,conn)
print(table)
```
